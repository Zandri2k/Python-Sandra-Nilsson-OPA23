```python
text_path = "../data/pikachu.txt"  #hämtar ut filer utanför sin egen map


with open(text_path, "r") as file:
    data_text = file.read()

print(data_text)
```

```python
data = data_text.split("\n")[1:-1]

data[0]
```

```python
eval(data[0])
```

```python
data2 = []
for point in data: 
    data2.append(eval(point)) #omvandlas till siffror

print(data2)
```

```python
data2[0]
```
